```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel(r'C:\Users\ronny\Desktop\project 1 diabetes\State pie.xlsx')
data.head()

```

```python
data.columns
```

```python
# Loop through each state in the data


# Loop through each row in the DataFrame
for index, row in data.iterrows():
    state = data.loc[index, 'State']
    medical_cost = data.loc[index, 'Medical_Costs']
    indirect_cost = data.loc[index, 'Indirect_Costs']
    
    # Create a pie chart
    #To show the original numbers of medical costs and indirect costs in the label of each pie chart
    labels = [f'Medical Costs:\n${medical_cost} billion', f'Indirect Costs: \n${indirect_cost} billion']
    label_props = {'weight': 'bold', 'fontsize': 8}
    sizes = [medical_cost, indirect_cost]
    colors = ['#1f77b4', '#ff7f0e']  # Blue for Medical Costs, Orange for Indirect Costs
    explode = (0.1, 0)  # Only "explode" the 1st slice (i.e., 'Medical Costs')
    
    plt.figure(figsize=(6, 6))  # Create a new figure with a specified size
    plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
            shadow=True, startangle=140, textprops=label_props)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title(f'State Level Cost Component for {state}\n', weight = 'bold', fontsize = 12)
    
    #  Save the figure  # Saves the pie chart as a PNG file named after the state
    plt.savefig(f'C:/Users/ronny/Desktop/project 1 diabetes/states cost pie graph/{state}_pie_chart.png')

    # To display the pie chart uncomment the next line
    plt.show()
    # Close the figure to free up memory
    plt.close()
```
